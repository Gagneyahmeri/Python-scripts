c) Muokataan seuraavaksi päällekkäiset rivit ja osa NaN-tiedoista pois. Tutki ensin Pandasin dokumentaatiosta funktioita päällekkäisten rivien ja nan-arvojen poistamiseen. Millä parametrilla asetat muokkauksen koskemaan koko dataframea, etkä palauta pelkkää näkymää? 

d) Tiputa datasta pois tuplarivit ja rivit, joilla sekä 'Source Network' että 'Destination Network' -tiedot ovat NaN (ei siis jompikumpi vaan molemmat ovat NaN).

```python
import pandas as pd

df_2022 = pd.read_csv('https://student.labranet.jamk.fi/~ajota/Datasets/cybersecurity_attacks_2022.csv', index_col=0)
df_2023 = pd.read_csv('https://student.labranet.jamk.fi/~ajota/Datasets/cybersecurity_attacks_2023_v2.csv', index_col=0)
```

```python
# Solu / vastaus 1.1 a) - datan yhdistäminen
# Tulosta yhdistetyn dataframen pituus/rivien määrä näkyviin ja kopioi Moodlen vastauskenttään.

df_2022_2023 = pd.concat([df_2022, df_2023])

len(df_2022_2023)
```

```python
# Solu / vastaus 1.1 b) - tuplarivien määrä
# Tulosta tuplarivien määrä näkyviin ja kopioi lukumäärä Moodlen vastauskenttään. (Pandasin dokumentaatiosta: '...the first occurrence is set on False and all others on True.' Mennään tässä tehtävässä perusoletusten mukaan.)

df_2022_2023.duplicated().sum()
```

```python
# Solu / vastaus 1.1 c)
# Tutki Pandasin dokumentaatiota / kurssimateriaaleja, millä parametrilla muokkaus tehdään alkuperäiseen dataframeen. Kirjoita parametrin nimi Moodlen vastauskenttään.

'inplace=True'
```

```python
# Solu / vastaus 1.1 d) - tuplien ja NaN-rivien tiputtaminen
# Kopioi Moodlen vastauskenttään muokatun dataframen rivien lukumäärä.

df_2022_2023.drop_duplicates(inplace=True)
df_2022_2023.dropna(subset=['Source Network', 'Destination Network'], how='all', inplace=True)

len(df_2022_2023)
```

```python
# Solu / vastaus 1.1 e) - uudelleen indeksointi ja tietojen järjestäminen
# Kopioi Moodlen vastauskenttään järjestetystä dataframesta viimeisen rivin 'Source IP Address'-tieto

df_2022_2023 = df_2022_2023.sort_values(by='Timestamp').reset_index(drop=True)

df_2022_2023.tail(1)
```

## Tietojen yhdistäminen - osa 2

Lisää edellisessä harjoituksessa muokattuun dataframeen df_2022_2023 tieto IP-osoitteiden maanosasta ja maan nimestä (continent_name ja country_name -tiedot dataframesta df_geo). 
- Lisää tiedot sekä lähde IP:lle että kohde IP:lle. Tavoitteena on, että tiedetään mistä mihin liikennettä on ollut.
- Dataframeihin df_2023 ja df_2022 on valmiiksi muodostettu sarakkeet Source Network ja Destination Network, joita voit hyödyntää yhdistämisessä.
- Mikäli yhdistämisen seurauksena sait aikaan päällekkäisiä tietoja sisältäviä sarakkeita, tiputa nämä pois.
- Nimeä uudet sarakkeet seuraavasti: Source IP Continent, Source IP Country, Destination IP Continent ja Destination IP Country.

Tulosta "Timestamp", "Source IP Address", "Destination IP Address", "Source Network", "Destination Network", "Source IP Continent", "Source IP Country", "Destination IP Continent", "Destination IP Country" -tiedot näkyviin.

```python
df_geo = pd.read_csv('https://student.labranet.jamk.fi/~ajota/Datasets/geoip2-ipv4.csv')
```

```python
# Solu / vastaus 1.2
# Kopioi Moodlen vastauskenttään sarakkeen 'Destination IP Country' viimeisen rivin maatieto.

import numpy as np

def get_geo_info(ip_network, df_geo):
    match = df_geo[df_geo['network'] == ip_network]
    return match[['continent_name', 'country_name']].values[0] if not match.empty else [np.nan]

df_2022_2023[['Source IP Continent', 'Source IP Country']] = df_2022_2023['Source Network'].apply(lambda x: get_geo_info(x, df_geo)).apply(pd.Series)
df_2022_2023[['Destination IP Continent', 'Destination IP Country']] = df_2022_2023['Destination Network'].apply(lambda x: get_geo_info(x, df_geo)).apply(pd.Series)

df3 = df_2022_2023[['Timestamp',
                    'Source IP Address',
                    'Destination IP Address',
                    'Source Network',
                    'Destination Network',
                    'Source IP Continent',
                    'Source IP Country',
                    'Destination IP Continent',
                    'Destination IP Country']]
df3.tail(1)
```
