## Uusien sarakkeiden muodostaminen - osa 2

Muodostetaan tässä kohdassa apusarake "network", jonka avulla harjoituksen viimeisessä kohdassa  saadaan lopulta yhdistettyä DataFrameen IP-osoitetta vastaava maatieto.

Tiedostossa geoip2-ipv4.csv on listattu IP-osoitteiden maanosia ja maita.

1. Lue data DataFrameen df_geo urlista https://student.labranet.jamk.fi/~ajota/Datasets/geoip2-ipv4.csv
2. Muodosta DataFrameen df_m uusi sarake "network", jossa on tieto mihin df_geo DataFramessa listattuihin verkkoihin IP kuuluu. Tiedostossa ei välttämättä ole tietoja kaikista verkoista. Käytä/aseta puuttuviksi arvoiksi Numpyn NaN-arvoja.
3. Tulosta dataframesta 5 ensimmäistä riviä näkyviin.

Voit lähteä ratkomaan tehtävää esim. seuraavasti, muutkin ratkaisut, joilla saat network-sarakkeen muodostettua DataFrameen ovat ok.
- Tallenna DataFramen df_geo "network"-arvot listaan
- Määrittele funktio, joka hoitaa tarkistuksen listan verkkojen ja IP:n välillä. Jos IP kuuluun verkon maskiin, funktio palauttaa ko. verkon.
- Tee tarkistus DataFramen df_m IP:ille.

Pisteytys: 1 p

Alla kuvakaappaus tavoitellusta lopputuloksesta:

![malli](https://student.labranet.jamk.fi/~ajota/Data-analytics/images/malli_add_network_column.png)

```python
# Luetaan maatiedot
df_geo = pd.read_csv('https://student.labranet.jamk.fi/~ajota/Datasets/geoip2-ipv4.csv')
```

```python
# Koodit tähän soluun
import numpy as np

network_list = df_geo['network'].tolist()

def find_network(ip):
    for network in network_list:
        try:
            if ip in ipaddress.IPv4Network(network):
                return network
        except ValueError:
            continue
    return np.nan

df_m['network'] = df_m['Source IP Address type(ip)'].apply(find_network)
```
