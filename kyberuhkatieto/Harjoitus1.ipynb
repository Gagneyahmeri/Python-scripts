## Pandas Series - alkioiden valinta, tunnusluvut ja suodattaminen

Seuraavassa solussa on tehty Pandasin series listasta "severity_level". Indeksien labeleiksi on asetettu päivämääriä kellonaikoineen.

a) Valitse serieksestä kaksi viimeisintä arvoa käyttäen indeksinumeroita.

b) Valitse serieksen arvot aikaväliltä 2024-01-02 08:00:00 - 2024-01-02 11:00:00 käyttäen indeksien labeleita.

c) Mitä arvoja serieksessä esiintyy? (Tulosta uniikit arvot näkyviin)

d) Valitse ne rivit, joilla arvo on "High".

e) Kuinka monta "High" tai "Medium" -arvoa serieksessä on?

Pisteytys: 0.3 p kustakin kohdasta, yht. 1.5 p.

```python
import pandas as pd

# Data listassa
severity_level = ['Low', 'Low', 'Low', 'Medium', 'Low', 'Medium', 'High', 'High', 'High', 'Medium', 'Medium', 'Low', 'Medium', 'Low', 'High',
       'Medium', 'High', 'Low', 'High']

# Indeksien luonti
indexes = pd.date_range(start='1/2/2024', freq='H', periods=19)

# Serieksen luonti
ser = pd.Series(data=severity_level, index=indexes)
```

```python
# Solu /vastaus 1.1 a)
# Valitse serieksestä kaksi viimeisintä arvoa käyttäen indeksinumeroita.

viimeiset = ser[-2:]
print(viimeiset)
```

```python
# Solu / vastaus 1.1 b)
# Valitse serieksestä "ser" arvot aikaväliltä 2024-01-02 08:00:00 - 2024-01-02 11:00:00 käyttäen indeksien labeleita, eli käyttäen suodatuksessa indeksien nimiä '2024-01-02 08:00:00' ja '2024-01-02 11:00:00'.

print(ser.loc['2024-01-02 08:00:00':'2024-01-02 11:00:00'])
```

```python
# Solu / vastaus 1.1 c)
# Mitä arvoja serieksessä esiintyy? (Selvitä millä funktiolla saat tulostettua serieksen uniikit arvot näkyviin. Tuloksena saat listan uniikeista arvoista.)

print(ser.unique())
```

```python
# Solu / vastaus 1.1 d)
# Valitse ne rivit, joilla arvo on "High".

print(ser[ser == 'High'])
```

```python
# Solu / vastaus 1.1 e)
# Kuinka monta "High" tai "Medium" -arvoa serieksessä on? Vastauksena on lukumäärä.

med_or_high = ser[(ser == 'High') | (ser == 'Medium')]
print(len(med_or_high))
```

## Pandas Series - alkioiden tunnusluvut ja suodattaminen jatkuu...

Alla on kaksi listaa, josta lista "anomaly_score" sisältää tietoliikenteen poikkeavuuden lukuarvona. Listassa "severity_level" on puolestaan luokittelu "anomaly_score" -listan arvot kolmeen luokkaan 'Low', 'Medium' ja 'High', jotka kuvaavat liikenteen poikkeavuuden tasoa. Tee listoista kaksi Seriestä (nimetty esim. sl ja ans). Voit ajatella että sinulla on kaksi taulukon saraketta, joista yhden tai molempien avulla saat suodatettua tietoja. Selvitä suodattamalla ja valmiiden funktioiden avulla seuraavat asiat:

a) Kuinka moni alkio Anomaly Scores -serieksessä on yli 50?

b) Kuinka monella alkiolla Anomaly Score -luku on yli 50 ja Severity Level -tieto on "High" tai "Medium"?

c) Mikä on Anomaly Scores -arvojen keskiarvo?

d) Mikä on muiden kuin "High"-arvojen maksimi (sl ei ole "High")?

e) Valitse Severity Level -serieksen indeksien 3 ja 4 arvoista 3 ensimmäistä kirjainta. Huomaa että indeksointi alkaa nollasta.


Pisteytys: 0.3 p kustakin kohdasta, yht. 1.5 p.

```python
anomaly_scores = [28.67, 51.5 , 87.42, 15.79,  0.52,  5.76, 31.55, 54.05, 56.34,
       16.51, 24.91, 86.07, 74.2 , 62.14, 72.65, 91.56, 43.83, 40.46, 44.75]

severity_level = ['Low', 'Low', 'Low', 'Medium', 'Low', 'Medium', 'High', 'High', 'High', 'Medium', 'Medium', 'Low', 'Medium', 'Low', 'High',
       'Medium', 'High', 'Low', 'High']

sl = pd.Series(severity_level) # Severity Level -series
ans = pd.Series(anomaly_scores) # Anomaly Scores -series
```

```python
# Solu / vastaus 1.2 a)
# Kuinka moni alkio Anomaly Scores -serieksessä on yli 50?

over50 = ans[ans > 50]

print(len(over50))
```

```python
# Solu / vastaus 1.2 b)
# Kuinka monella alkiolla Anomaly Score -luku on yli 50 ja Severity Level -tieto on "High" tai "Medium"?

over50_med_high = ans[(ans > 50) & (sl != 'Low')]

print(len(over50_med_high))
```

```python
# Solu / vastaus 1.2 c)
# Mikä on Anomaly Scores -arvojen keskiarvo keskiarvo? Käytä valmista funktioita.

average = round(ans.mean(), 2)

print(average)
```

```python
# Solu / vastaus 1.2 d)
# Mikä on muiden kuin "High"-arvojen maksimi (sl ei ole "High")?

maksimi = ans[sl != 'High'].max()

print(maksimi)
```

```python
# Solu / vastaus 1.2 e)
# Suodata ensin indeksien perusteella (indeksit 3 ja 4) oikeat arvot ja sitten valitse arvoista 'Medium' ja 'Low' ensimmäiset kolme kirjainta (str).

idx = sl[3:5]

med_high = idx[sl != 'High']

print(med_high.str[:3])
```
